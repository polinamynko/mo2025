### Параметрические графики
$$
x = \sin t/t,
\quad
y = \cos t/t
$$
$x = \frac{\sin t}{t},\ y = \frac{\cos t}{t}$

```python
t = np.linspace(1.5, 32, 200)
plt.axis('equal')
plt.title(r'$x = \frac{\sin t}{t},\ y = \frac{\cos t}{t}$')
x = np.sin(t)/t
y = np.cos(t)/t
plt.plot(x, y)
```

## Stacked normalized (percentage) bar plot

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'pre_score': [4, 24, 31, 2, 3],
        'mid_score': [25, 94, 57, 62, 70],
        'post_score': [5, 43, 23, 23, 51]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'pre_score', 'mid_score', 'post_score'])

f, ax = plt.subplots(1, figsize=(10,5))

bar_width = 1

bar_l = [i for i in range(len(df['pre_score']))]

tick_pos = [i+(bar_width/2) for i in bar_l]

totals = [i+j+k for i,j,k in zip(df['pre_score'], df['mid_score'], df['post_score'])]
pre_rel = [i / j * 100 for  i,j in zip(df['pre_score'], totals)]
mid_rel = [i / j * 100 for  i,j in zip(df['mid_score'], totals)]
post_rel = [i / j * 100 for  i,j in zip(df['post_score'], totals)]

ax.bar(bar_l, pre_rel,                  label='Pre Score', alpha=0.9, color='#019600', width=bar_width, edgecolor='white')
ax.bar(bar_l, mid_rel,  bottom=pre_rel, label='Mid Score', alpha=0.9, color='#3C5F5A', width=bar_width, edgecolor='white')
ax.bar(bar_l, post_rel, bottom=[i+j for i,j in zip(pre_rel, mid_rel)], label='Post Score',alpha=0.9,
       color='#219AD8', width=bar_width,edgecolor='white')

plt.xticks(tick_pos, df['first_name'])
ax.set_ylabel("Percentage")
ax.set_xlabel("")
ax.margins(0);

plt.show()
```

## Круговая диаграмма (Pie chart)

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips.day = tips.day.astype('category')  # Опционально, если нужно для других задач

f, ax = plt.subplots(1, figsize=(7, 7))

# Правильный способ: группируем по 'day' и суммируем только 'tip'
tips.groupby('day')['tip'].sum().plot(
    kind='pie',
    ax=ax,
    shadow=True,
    startangle=90,
    colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'],
    autopct='%1.1f%%'
)

ax.set_title('Чаевые по дням недели', fontsize=14)
ax.set_ylabel('')
plt.show()
```

## Задание 1
Нужно построить 4 гистограммы (histogram) на 30 карманов (бинов) главных 4 характеристик покемонов:
- HP - здоровья
- Attack - уровня атаки
- Defense - уровня защиты
- Speed - скорости
Построить их надо на одном графике с 4 подграфиками (axes). Нужно использовать параметр figsize.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/pokemon_full.csv'
data = pd.read_csv(url)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Гистограмма для HP (здоровья)
axes[0, 0].hist(data['HP'], bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Распределение HP (здоровья)')
axes[0, 0].set_xlabel('HP')
axes[0, 0].set_ylabel('Частота')

# Гистограмма для Attack (атаки)
axes[0, 1].hist(data['Attack'], bins=30, color='salmon', edgecolor='black')
axes[0, 1].set_title('Распределение Attack (атаки)')
axes[0, 1].set_xlabel('Attack')
axes[0, 1].set_ylabel('Частота')

# Гистограмма для Defense (защиты)
axes[1, 0].hist(data['Defense'], bins=30, color='lightgreen', edgecolor='black')
axes[1, 0].set_title('Распределение Defense (защиты)')
axes[1, 0].set_xlabel('Defense')
axes[1, 0].set_ylabel('Частота')

# Гистограмма для Speed (скорости)
axes[1, 1].hist(data['Speed'], bins=30, color='gold', edgecolor='black')
axes[1, 1].set_title('Распределение Speed (скорости)')
axes[1, 1].set_xlabel('Speed')
axes[1, 1].set_ylabel('Частота')

plt.tight_layout()    # Настройка отступов между подграфиками

plt.show()     # Показать график
```

## Задание 2
Нужно выбрать несколько полей и построить диаграммы:
- круговая
- гистрограмма
- график
- ящик с усами
- ViolinPlot

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/pokemon_full.csv'
df = pd.read_csv(url)

type_counts = df['Type_1'].value_counts()

# 1. Построение круговой диаграммы
plt.figure(figsize=(10, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение покемонов по типу')
plt.axis('equal')
plt.show()

# 2. Построение гистограммы
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))     # Создаем subplot (3 гистограммы в одном окне)

# Гистограмма Attack
axes[0].hist(df['Attack'], bins=20, color='red', edgecolor='black')
axes[0].set_title('Распределение атаки (Attack)')
axes[0].set_xlabel('Значение атаки')
axes[0].set_ylabel('Количество покемонов')

# Гистограмма Defense
axes[1].hist(df['Defense'], bins=20, color='blue', edgecolor='black')
axes[1].set_title('Распределение защиты (Defense)')
axes[1].set_xlabel('Значение защиты')

# Гистограмма Speed
axes[2].hist(df['Speed'], bins=20, color='green', edgecolor='black')
axes[2].set_title('Распределение скорости (Speed)')
axes[2].set_xlabel('Значение скорости')

plt.tight_layout()  # Чтобы графики не накладывались
plt.show()

# 3. Построение точечного графика
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['Attack'], df['Defense'],
                     c=pd.factorize(df['Type_1'])[0],
                     cmap='tab20', alpha=0.7)
plt.title('Зависимость Attack от Defense (по типам)')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.colorbar(scatter, label='Тип покемона (Type_1)')
plt.grid(linestyle='--', alpha=0.5)
plt.show()

# 4. Построение графика ящик с усами
plt.figure(figsize=(8, 6))
df.boxplot(column='HP', by='isLegendary', grid=False,
           patch_artist=True,
           boxprops=dict(facecolor='lightblue'),
           medianprops=dict(color='red'))
plt.title('Распределение HP у легендарных и обычных покемонов')
plt.suptitle('')  # Убираем автоматический заголовок
plt.xlabel('Легендарный?')
plt.ylabel('HP (здоровье)')
plt.xticks([1, 2], ['Обычный', 'Легендарный'])
plt.show()

# 5. Построение графика ViolinPlot
plt.figure(figsize=(8, 6))

sns.violinplot(x='isLegendary', y='Defense', data=df,
               palette='pastel', inner='quartile')  # inner='quartile' показывает квантили

plt.title('Защита (Defense) у легендарных и обычных покемонов')
plt.xlabel('Легендарный покемон?')
plt.ylabel('Защита (Defense)')
plt.xticks([0, 1], ['Нет', 'Да'])
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```
