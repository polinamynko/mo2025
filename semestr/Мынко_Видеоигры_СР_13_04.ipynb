#### 23. Преобразуйте все категориальные поля в числовые с помощью LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

# 1. Выделяем категориальные колонки (тип object и category)
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 2. Создаем копию DataFrame для преобразований
df_encoded = df.copy()

# 3. Инициализируем LabelEncoder и преобразуем каждую категориальную колонку
le = LabelEncoder()

for col in cat_cols:
    # Заполняем пропуски специальным значением перед кодированием
    df_encoded[col] = df_encoded[col].fillna('MISSING')

    # Преобразуем в строковый тип (на случай mixed-типов)
    df_encoded[col] = df_encoded[col].astype(str)

    # Применяем LabelEncoder
    df_encoded[col] = le.fit_transform(df_encoded[col])

    # Сохраняем соответствие меток для обратного преобразования
    print(f"\nКолонка: {col}")
    print("Соответствие значений:")
    for i, class_ in enumerate(le.classes_):
        print(f"{i}: {class_}")

# 4. Проверяем результат
print("\nПервые 5 строк преобразованных данных:")
print(df_encoded.head())

print("\nТипы данных после преобразования:")
print(df_encoded[cat_cols].dtypes)
```
